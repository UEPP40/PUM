```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import csv
```

```python
path = "/content/sample_data/GroceryStoreDataSet.csv"
# reader = csv.reader(path)
# dataset = list(reader)

# print(dataset)

dataset = pd.read_csv(path, sep=",", header = None)
# dataset = dataset.apply(lambda s:s.str.replace('"', ""))
dataset.head()
```

```python
transactions = []
items = dataset[0].values
print(items)
```

```python
for i in range (0,len(items)):
  transactions.append(items[i].split(","))
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions) ## !!dataset – lista list
```

```python
te_ary.astype("int")
```

```python
te.columns_
```

```python
pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(pd.DataFrame(te_ary, columns=te.columns_), min_support=0.1, use_colnames=True)
conf = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

conf.sort_values(by='confidence', ascending=False)
print(conf)
```
