```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import math
from sklearn.model_selection import train_test_split


df = pd.read_csv("train.csv")
# df.head(100)
```

```python
df.shape
```

```python
duplicates = df.duplicated(subset=['ID'])
if duplicates.any():
    print("Znaleziono duplikaty w kolumnie 'ID'.")
    print(df[duplicates])
else:
    print("Brak duplikatów w kolumnie 'ID'.")
```

```python
df.drop_duplicates
```

```python
FeaturesToConvert = ['Age', 'Annual_Income', 'Num_of_Loan', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Outstanding_Debt', 'Amount_invested_monthly', 'Monthly_Balance']

for feature in FeaturesToConvert:
   df[feature] = df[feature].str.strip('-_')

for feature in FeaturesToConvert:
   df[feature] = df[feature].replace({'':np.nan})

for feature in FeaturesToConvert:
   df[feature] = df[feature].astype('float64')
```

```python
df['Monthly_Inhand_Salary']= df['Monthly_Inhand_Salary'].fillna(method='pad')
df['ID'] = df['ID'].fillna('Brak informacji')
df['Customer_ID'] = df['Customer_ID'].fillna('Brak informacji')
df['Month'] = df['Month'].fillna('Brak informacji')
df['Name'] = df['Name'].fillna('Nieznane')
age_mean = df[(df['Age'] <= 100) & (df['Age'] >= 1)]['Age'].mean()
df['Age'] = df['Age'].fillna(age_mean)
df.loc[df['Age'] >= 100, 'Age'] = age_mean
df.loc[df['Age'] < 1, 'Age'] = age_mean
df['SSN'] = df['SSN'].fillna('Brak informacji')
df['Occupation'] = df['Occupation'].fillna('Brak informacji')
df['Annual_Income'] = df['Annual_Income'].fillna(method='pad')
df['Num_Bank_Accounts'] = df['Num_Bank_Accounts'].fillna(0)
df['Num_Credit_Card'] = df['Num_Credit_Card'].fillna(0)
df['Interest_Rate'] = df['Interest_Rate'].fillna(0)
df['Num_of_Loan'] = df['Num_of_Loan'].fillna(0)
df['Type_of_Loan'] = df['Type_of_Loan'].fillna('Nieokreślony')
df['Delay_from_due_date'] = df['Delay_from_due_date'].fillna(0)
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].fillna(0)
df['Changed_Credit_Limit'] = df['Changed_Credit_Limit'].fillna(0)
df['Num_Credit_Inquiries'] = df['Num_Credit_Inquiries'].fillna(0)
df['Credit_Mix'] = df['Credit_Mix'].fillna('Nieznane')
df['Outstanding_Debt'] = df['Outstanding_Debt'].fillna(0)
df['Credit_Utilization_Ratio'] = df['Credit_Utilization_Ratio'].fillna(0)
df['Credit_History_Age'] = df['Credit_History_Age'].fillna('Brak historii kredytowej')
df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].fillna('Brak informacji')
df['Total_EMI_per_month'] = df['Total_EMI_per_month'].fillna(method='pad')
df['Amount_invested_monthly'] = df['Amount_invested_monthly'].fillna(method='pad')
df['Payment_Behaviour'] = df['Payment_Behaviour'].fillna('Brak informacji')
df['Monthly_Balance'] = df['Monthly_Balance'].fillna(method='pad')
df['Credit_Score'] = df['Credit_Score'].fillna('Brak informacji')
```

```python
# tworzymy obiekt encodera czyli po prostu tworzymy nowy obiekt klasy LabelEncoder(ktora jest zdefiniowana w kodzie biblioteki) ktory przypisujemy do zmiennej le
# dokumentacja Encodera - https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html
# kod klasy LabelEncoder - https://github.com/scikit-learn/scikit-learn/blob/5c4aa5d0d/sklearn/preprocessing/_label.py#L34
le = LabelEncoder()

# Jako przykład zakodujmy wartości dla kolumny Age, zwrócone zostają wartości zakodowane (dla ambitnych podrzucam kod funkcji i kod funkcji wykorzystywanej w fit_transform https://github.com/scikit-learn/scikit-learn/blob/5c4aa5d0d/sklearn/preprocessing/_label.py#L101  https://github.com/scikit-learn/scikit-learn/blob/5c4aa5d0d90ba66247d675d4c3fc2fdfba3c39ff/sklearn/utils/_encode.py#L10)
df.Age = le.fit_transform(df.Age)
# kodowanie dla kilku kolumn odbywa sie podobnie tylko najpierw musimy zadefiniowac tablice z nazwami kolumn które mają przejść transformację
# dałam tylko kilka dla przykładu działania
columns_to_code = ['Occupation','Annual_Income','Monthly_Balance', 'Credit_Score']

# dzięki apply możemy za jednym wywołaniem kodu wykonać daną operację na każdym elemencie z listy
df[columns_to_code] = df[columns_to_code].apply(LabelEncoder().fit_transform)
```

```python
def convert_to_years(text):
    if isinstance(text, str):
        parts = text.split()
        if parts[0].isdigit() and parts[3].isdigit():  # Sprawdzenie czy pierwszy i czwarty element jest liczbą
            year = int(parts[0])
            month = int(parts[3])
            credit_age_numeric = year + month / 12
            return credit_age_numeric
        else:
            return None  # Zwrócenie None dla niepoprawnych wartości
    else:
        return text

df["Credit_History_Age"] = df["Credit_History_Age"].apply(lambda x: convert_to_years(x))
```

```python
df["Credit_History_Age"] = df["Credit_History_Age"].fillna(0)
df["Credit_History_Age"] = df["Credit_History_Age"].astype(float)
df.head(50)
```
