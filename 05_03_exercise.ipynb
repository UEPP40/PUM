```python
import pandas as pd
df1 = pd.read_csv('train.csv')
```

```python
df1.shape
```

```python
df1.drop_duplicates()
```

```python
df1 = df1.drop(columns = ["ID", "Customer_ID", "Name", "SSN"])
```

```python
df1['Age'] = df1['Age'].str.replace('_','')
```

```python
df1['Annual_Income'] = df1['Annual_Income'].str.replace('_','')
```

```python
df1['Monthly_Balance'] = df1['Monthly_Balance'].str.replace('_','')
```

```python
df1['Num_of_Loan'] = df1['Num_of_Loan'].str.replace('_','')
```

```python
df1['Outstanding_Debt'] = df1['Outstanding_Debt'].str.replace('_','')
```

```python
df1['Amount_invested_monthly'] = df1['Amount_invested_monthly'].str.replace('_','')
```

```python
df1[["Age"]] = df1[["Age"]].apply(pd.to_numeric)
```

```python
df1 = df1.astype({'Annual_Income' : 'float64', 'Monthly_Balance' : 'float64', 'Num_of_Loan' : 'int64', 'Outstanding_Debt' : 'float64', 'Amount_invested_monthly' : 'float64'})
```

```python
df1.dtypes
```

```python
df1['Monthly_Inhand_Salary'] = df1['Monthly_Inhand_Salary'].fillna(method='pad')
```

```python
df1['Month'] = df1['Month'].fillna(method='bfill')
```

```python
df1['Age'] = df1['Age'].clip(lower = 0)
```
