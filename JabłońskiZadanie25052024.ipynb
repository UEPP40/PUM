```python
import random
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
#1
df = pd.read_csv("GroceryStoreDataSet.csv", header=None)
df.head()
```

```python
transactions_list = [row.split(',') for row in df[0]]
te = TransactionEncoder()
te_ary = te.fit(transactions_list).transform(transactions_list)
te_ary
te_ary.astype("int")
```

```python
df_data = pd.DataFrame(te_ary, columns = te.columns_)
df_data
```

```python
supp = apriori(df_data, min_support = 0.15, use_colnames = True).sort_values(by = 'support', ascending = False)
supp
```

```python
conf = conf = association_rules(supp, metric="confidence", min_threshold=0.05).sort_values(by = 'confidence', ascending = False)
conf
```

```python
conf1 = association_rules(supp, metric='zhangs_metric', min_threshold=0.15).sort_values(by='zhangs_metric', ascending=False)
conf1
```

```python
M = df['Id'].nunique()
N = df['movie'].nunique()
user_mapper = dict(zip(np.unique(df["Id"]), list(range(M))))
movie_mapper = dict(zip(np.unique(df["movie"]), list(range(N))))
user_inv_mapper = dict(zip(list(range(M)), np.unique(df["Id"])))
movie_inv_mapper = dict(zip(list(range(N)), np.unique(df["movie"])))
user_index = [user_mapper[i] for i in df['Id']]
item_index = [movie_mapper[i] for i in df['movie']]
X = csr_matrix((df["rating"], (user_index, item_index)), shape=(M,N))
```

```python
n_total = X.shape[0]*X.shape[1]
n_ratings = X.nnz #scipy.sparse.csr_matrix.nnz no of stored values, including explicit 0
sparsity = n_ratings/n_total
sparsity_p = round(sparsity*100,2)
print(f"Matrix: {sparsity_p}%")
n_ratings_per_user = X.getnnz(axis=1)
```

```python
def find_similar_movies(movie_id, X, movie_mapper, movie_inv_mapper, k, metric='cosine'):
    X = X.T
    neighbour_ids = []
    movie_ind = movie_mapper[movie_id]
    movie_vec = X[movie_ind]
    if isinstance(movie_vec, (np.ndarray)):
        movie_vec = movie_vec.reshape(1,-1)

    kNN = NearestNeighbors(n_neighbors=k+1, algorithm="brute", metric=metric)
    kNN.fit(X)
    neighbour = kNN.kneighbors(movie_vec, return_distance=False)
    for i in range(0,k):
        n = neighbour.item(i)
        neighbour_ids.append(movie_inv_mapper[n])
    neighbour_ids.pop(0)
    return neighbour_ids
```

```python
similar_movies = find_similar_movies(movie_id=1,X=X,movie_mapper=movie_mapper,movie_inv_mapper=movie_inv_mapper, k=10)
similar_movies

movie_names = df.loc[df['movieId'].isin(similar_movies), 'title'].tolist()
set(movie_names)
```

```python
from surprise import Dataset, Reader, SVD, KNNBasic
from surprise.model_selection import GridSearchCV, cross_validate

data_reader = Reader(rating_scale=(0.5, 5.0))
ratings_data = Dataset.load_from_df(df[['userId', 'movieId', 'rating']], data_reader)

svd_param_grid = {'n_factors': [15, 30], 'n_epochs': [20, 30], 'lr_all': [0.005, 0.010]}
svd_grid_search = GridSearchCV(SVD, svd_param_grid, measures=['rmse', 'mae'], cv=5)
svd_grid_search.fit(ratings_data)

random_neighbors = random.randint(5, 40)
print(f"Randomly selected number of neighbors: {random_neighbors}")

knn_param_grid = {'k': [random_neighbors], 'sim_options': {'name': ['msd', 'cosine'], 'user_based': [False]}}
knn_grid_search = GridSearchCV(KNNBasic, knn_param_grid, measures=['rmse', 'mae'], cv=5)
knn_grid_search.fit(ratings_data)

best_svd_model = svd_grid_search.best_estimator['rmse']
best_knn_model = knn_grid_search.best_estimator['rmse']

cv_svd_results = cross_validate(best_svd_model, ratings_data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
cv_knn_results = cross_validate(best_knn_model, ratings_data, measures=['RMSE', 'MAE'], cv=5, verbose=True)

print("SVD:", cv_svd_results)
print("kNN:", cv_knn_results)
```
