```python
# Funkcja do konwersji wieku historii kredytowej na liczbe miesiecy
def convert_age_to_months(age_str):
    if pd.isna(age_str):
        return None
    parts = age_str.split(' ')
    years = int(parts[0]) if parts[0].isdigit() else 0
    months = int(parts[3]) if len(parts) > 3 and parts[3].isdigit() else 0
    return years * 12 + months

# Nadpisanie kolumny 'Credit_History_Age' przekonwertowanymi wartościami
df['Credit_History_Age'] = df['Credit_History_Age'].apply(convert_age_to_months)

# Usunięcie wierszy z wartościami NaN w kolumnie 'Credit_History_Age'
df = df.dropna(subset=['Credit_History_Age'])

df[['Credit_History_Age']].head()
```

```python
df['Credit_History_Age']= df['Credit_History_Age'].astype(float)
```

```python
# Usunięcie niepożądanych znaków i konwersja 'Annual_Income' na typ float
df['Annual_Income'] = df['Annual_Income'].str.replace('_', '').astype(float)

# Ponowne obliczenie IQR dla 'Annual_Income' po oczyszczeniu danych
Q1 = df['Annual_Income'].quantile(0.25)
Q3 = df['Annual_Income'].quantile(0.75)
IQR = Q3 - Q1

# Usunięcie zmiennych odstających z 'Annual_Income'
df = df.drop(df.loc[df['Annual_Income'] > (Q3 + 1.5 * IQR)].index)
df = df.drop(df.loc[df['Annual_Income'] < (Q1 - 1.5 * IQR)].index)

df[['Annual_Income']].describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Utworzenie instancji scaler'a (tutaj użyjemy MinMaxScaler)
scaler = MinMaxScaler()

# Lista kolumn do ustandaryzowania/znormalizowania
col_float = [
    'Age', 'Annual_Income', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
    'Outstanding_Debt', 'Credit_History_Age', 'Total_EMI_per_month', 'Monthly_Balance'
]

# Przed normalizacją konwersja typów kolumn na float, gdzie to konieczne
df['Age'] = df['Age'].astype(float)
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].astype(float)
df['Outstanding_Debt'] = df['Outstanding_Debt'].astype(float)
df['Monthly_Balance'] = df['Monthly_Balance'].astype(float)

# Normalizacja wybranych kolumn
for i in col_float:
    df[i] = scaler.fit_transform(df[[i]])

df.head()
```

```python
# Usunięcie niepożądanych znaków z kolumn, które będą normalizowane
for col in col_float:
    if df[col].dtype == 'object' or df[col].dtype == 'string':
        df[col] = df[col].str.replace('_', '').astype(float)

# Ponowna próba normalizacji wybranych kolumn
for i in col_float:
    df[i] = scaler.fit_transform(df[[i]])

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Wybranie kolumn numerycznych
numerical_cols = [
    'Age', 'Annual_Income', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
    'Outstanding_Debt', 'Credit_History_Age', 'Total_EMI_per_month', 'Monthly_Balance'
]

# Rysowanie heatmapy korelacji między wybranymi cechami numerycznymi
plt.figure(figsize=(20, 18))
sns.heatmap(df[numerical_cols].corr(), annot=True, linewidths=0.1, cmap='Blues')
plt.title('Numerical Features Correlation')
plt.show()
```

```python
# Ponieważ 'Credit_Score' jest zmienną kategoryczną, nie możemy bezpośrednio użyć jej
# w standardowej korelacji Pearsona z innymi zmiennymi numerycznymi.
# Zamiast tego, możemy zakodować 'Credit_Score' numerycznie i pokazać korelację
# między tym nowym kodowaniem a innymi zmiennymi numerycznymi.

# Kodowanie 'Credit_Score' jako wartości numeryczne
credit_score_mapping = {label: idx for idx, label in enumerate(df['Credit_Score'].unique())}
df['Credit_Score_Encoded'] = df['Credit_Score'].map(credit_score_mapping)

# Wybranie kolumn numerycznych plus zakodowany 'Credit_Score'
all_numerical_cols = numerical_cols + ['Credit_Score_Encoded']

# Rysowanie heatmapy korelacji między wszystkimi cechami numerycznymi i zakodowanym 'Credit_Score'
plt.figure(figsize=(22, 20))
sns.heatmap(df[all_numerical_cols].corr(), annot=True, linewidths=0.1, cmap='Blues', fmt=".2f")
plt.title('Numerical Features Correlation with Encoded Credit Score')
plt.show()
```

```python
# Ponieważ wystąpił błąd, spróbujmy jeszcze raz wykonać poprzedni kod w poprawny sposób.

# Utworzenie DataFrame z rankingiem korelacji cech względem 'Credit_Score_Encoded'
correlation_with_credit_score = pd.DataFrame(abs(df.corr()['Credit_Score_Encoded'].drop('Credit_Score_Encoded') * 100).sort_values(ascending=False))

# Wyświetlenie rankingu korelacji w formie wykresu słupkowego
correlation_with_credit_score.plot.bar(figsize=(10, 8), legend=False)
plt.title('Feature Correlation with Encoded Credit Score (%)')
plt.show()

# Usunięcie dwóch najmniej skorelowanych cech
least_correlated_features = correlation_with_credit_score.tail(2).index
df_reduced = df.drop(columns=least_correlated_features)

# Zapisanie zmodyfikowanego DataFrame jako 'zbiór nr 1'
zbiór_nr_1 = df_reduced

least_correlated_features
```
