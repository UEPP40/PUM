```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('train.csv')
print("All:", df_train.shape)
df_train.drop_duplicates(subset="ID", inplace=True)
print("Without duplicates:", df_train)
```

```python
df_train.dtypes
```

```python
df_train['Age'] = pd.to_numeric(df_train['Age'], errors='coerce').astype('Int64')
df_train['Month'] = pd.to_numeric(df_train['Month'], errors='coerce').astype('string')
df_train['Annual_Income'] = pd.to_numeric(df_train['Annual_Income'], errors='coerce').astype('float64')
df_train['Num_of_Loan'] = pd.to_numeric(df_train['Num_of_Loan'], errors='coerce').astype('Int64')
df_train['Num_of_Delayed_Payment'] = pd.to_numeric(df_train['Num_of_Delayed_Payment'], errors='coerce').astype('Int64')
df_train['Changed_Credit_Limit'] = pd.to_numeric(df_train['Changed_Credit_Limit'], errors='coerce').astype('float64')
df_train['Outstanding_Debt'] = pd.to_numeric(df_train['Outstanding_Debt'], errors='coerce').astype('float64')
df_train['Amount_invested_monthly'] = pd.to_numeric(df_train['Amount_invested_monthly'], errors='coerce').astype('float64')
df_train['Monthly_Balance'] = pd.to_numeric(df_train['Monthly_Balance'], errors='coerce').astype('float64')
```

```python
df_train.dtypes
```

```python
df_train = df_train.drop(["ID", "Customer_ID", "Name", "SSN"], axis=1)
```
