```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.preprocessing import MinMaxScaler,StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split

df = pd.read_csv("train.csv")
df.drop_duplicates(subset="ID", inplace=True)

df.drop(columns=['ID','Customer_ID','SSN','Name'], inplace=True)

df['Credit_History_Age'] = (pd.to_numeric(df["Credit_History_Age"].str.split(" ", expand=True)[0], errors='coerce') * 12 +
                            pd.to_numeric(df["Credit_History_Age"].str.split(" ", expand=True)[3], errors='coerce'))

FeaturesToConvert = ['Age', 'Annual_Income', 'Num_of_Loan', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Outstanding_Debt', 'Amount_invested_monthly',
                     'Monthly_Balance']

for feature in FeaturesToConvert:
   df[feature] = df[feature].str.strip('-_')
   df[feature] = df[feature].replace({'':np.nan})
   df[feature] = df[feature].astype('float64')

df['Monthly_Inhand_Salary']= df['Monthly_Inhand_Salary'].fillna(method='pad')


age_mean = df[(df['Age'] <= 120) & (df['Age'] >= 1)]['Age'].mean()
df['Age'] = df['Age'].fillna(age_mean)

df.loc[df['Age'] > 120, 'Age'] = age_mean
df.loc[df['Age'] < 1, 'Age'] = age_mean

df['Month'] = df['Month'].fillna('No info')
df['Occupation'] = df['Occupation'].fillna('No info')
df['Type_of_Loan'] = df['Type_of_Loan'].fillna('No info')
df['Credit_Mix'] = df['Credit_Mix'].fillna('No info')
df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].fillna('No info')
df['Payment_Behaviour'] = df['Payment_Behaviour'].fillna('No info')
df['Credit_Score'] = df['Credit_Score'].fillna('No info')

df['Annual_Income'] = df['Annual_Income'].fillna(method='pad')
df['Total_EMI_per_month'] = df['Total_EMI_per_month'].fillna(method='pad')
df['Amount_invested_monthly'] = df['Amount_invested_monthly'].fillna(method='pad')
df['Monthly_Balance'] = df['Monthly_Balance'].fillna(method='pad')

df['Credit_History_Age'] = df['Credit_History_Age'].fillna(0)
df['Num_Bank_Accounts'] = df['Num_Bank_Accounts'].fillna(0)
df['Num_Credit_Card'] = df['Num_Credit_Card'].fillna(0)
df['Interest_Rate'] = df['Interest_Rate'].fillna(0)
df['Num_of_Loan'] = df['Num_of_Loan'].fillna(0)
df['Delay_from_due_date'] = df['Delay_from_due_date'].fillna(0)
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].fillna(0)
df['Changed_Credit_Limit'] = df['Changed_Credit_Limit'].fillna(0)
df['Num_Credit_Inquiries'] = df['Num_Credit_Inquiries'].fillna(0)
df['Outstanding_Debt'] = df['Outstanding_Debt'].fillna(0)
df['Credit_Utilization_Ratio'] = df['Credit_Utilization_Ratio'].fillna(0)


columns_to_code = ['Occupation','Annual_Income','Monthly_Balance', 'Credit_Score']

le = LabelEncoder()
df[columns_to_code] = df[columns_to_code].apply(LabelEncoder().fit_transform)
df.Age = le.fit_transform(df.Age)

df.head(10)
```
