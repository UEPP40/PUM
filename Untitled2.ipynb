```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import median_absolute_error
```

```python
df = pd.read_csv("Salary Data (2).csv")
```

```python
age_mean = df[(df['Age'] <= 100) & (df['Age'] >= 1)]['Age'].mean()
df.loc[df['Age'] >= 100, 'Age'] = age_mean
df.loc[df['Age'] < 1, 'Age'] = age_mean
```

```python
df['Age'] = df['Age'].fillna('age_mean')
df['Gender'] = df['Gender'].fillna(method='pad')
df['Education Level'] = df['Education Level'].fillna(method='pad')
df['Job Title'] = df['Job Title'].fillna(method='pad')
df['Years of Experience'] = df['Years of Experience'].fillna(method='pad')
df['Salary'] = df['Salary'].fillna(method='pad')
```

```python
df['Age'] = df['Age'].astype(int)
df['Years of Experience'] = df['Years of Experience'].astype(int)
```

```python
le = LabelEncoder()

columns_to_code = ['Age', 'Gender', 'Education Level', 'Job Title', 'Years of Experience', 'Salary']
df[columns_to_code] = df[columns_to_code].apply(LabelEncoder().fit_transform)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['Age','Gender','Education Level', 'Job Title', 'Years of Experience']], df['Salary'], test_size=0.33, random_state=42)
```

```python
reg = LinearRegression().fit(x_train, y_train)
y_pred = reg.predict(x_test)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
median_absolute_error(y_test, y_pred)
```
