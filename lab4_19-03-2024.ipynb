```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
```

```python
#przeczytaj dane z pliku csv
df = pd.read_csv("/content/sample_data/Salary Data (2).csv", sep = "," , encoding= 'utf-8')

df.info()
df.head(10)
```

```python
# enkoder płci, poziomu edukacji i nazwy zawodu
le = LabelEncoder()

columns = ["Gender", "Education Level", "Job Title"]
df[columns] = df[columns].apply(LabelEncoder().fit_transform)

df.dropna(inplace=True)

df.info()
df.head(10)
```

```python
#podział danych na uczące i testowe
train_df, test_df = train_test_split(df ,test_size=0.3, random_state=40)

X_train = train_df[["Age", "Gender", "Education Level", "Job Title", "Years of Experience"]]
Y_train = train_df["Salary"]
X_test = test_df[["Age", "Gender", "Education Level", "Job Title", "Years of Experience"]]
Y_test = test_df["Salary"]
```

```python
reg = LinearRegression().fit(X_train,Y_train)
reg.score(X_train,Y_train)
reg.coef_
reg.intercept_
Y_predic = reg.predict(X_test)

Y_true = Y_test
```

Średni błąd bezwzględny to średnia różnica między obserwacjami (wartościami rzeczywistymi) a wynikami modelu (prognozami). Znak tych różnic jest ignorowany, aby nie dochodziło do anulowania wartości dodatnich i ujemnych.

https://insidelearningmachines.com/mean_absolute_error/

```python
print ("Regresja liniowa | Średni błąd bezwzględny: ",  mean_absolute_error(Y_true,Y_predic))
```

Mediana błędu bezwględnego to mediana różnicy między obserwacjami (wartościami rzeczywistymi) a wynikami modelu (prognozami). Znak tych różnic jest ignorowany, aby nie dochodziło do anulowania wartości dodatnich i ujemnych.

Główną zaletą korzystania z tej metryki jest jej duża odporność na wartości odstające. Kontrastuje to z innymi metrykami, takimi jak średni błąd bezwzględny lub średni błąd kwadratowy.

https://insidelearningmachines.com/median_absolute_error/

```python
print("Regresja liniowa | Mediana błędu bezwględnego: ",median_absolute_error(Y_true,Y_predic))
```

Współczynnik determinacji jest miarą służącą do oceny jakości dopasowania modelu regresji liniowej. Przyjmuje on wartości z przedziału [0; 1]. Optymalnym wynikiem byłoby R^2 = 1,0, wskazując, że nasz model jest w stanie uwzględnić całą zmienność danych. Ponieważ model może być arbitralnie zły, nie ma dolnej granicy dla R^2

https://insidelearningmachines.com/coefficient_of_determination/

```python
print("Regresja liniowa | Współczynnik determinacj: ", r2_score(Y_true,Y_predic))
```
